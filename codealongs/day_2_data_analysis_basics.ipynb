```python
rpg_ps3 = df.loc[(df["Platform"] == "PS3") & (df["Genre"] == "Role-Playing")].reset_index(drop=True)
```

```python
rpg_ps3
```

```python
action = df["Genre"] == "Action"
shooter = df["Genre"] == "Shooter"


df.loc[action | shooter]
```

```python
since_2010 = df["Year"] >= 2010

df.loc[(since_2010) & (action | shooter)]
```

```python
# isin()

top_3_publishers = df.Publisher.isin(["Electronic Arts", "Activision", "Namco Bandai Games"])
```

```python
df.loc[top_3_publishers]
```

```python
# between
between_2M_10M = df["Global Sales"].between(2,10)

y_90s = df["Year"].between(1990,1999)
y_00s = df["Year"].between(2000,2009)
y_10s = df["Year"].between(2010,2019)
```

```python
df.loc[between_2M_10M & y_90s, "Genre"].value_counts(normalize=True)
```

```python
df.loc[between_2M_10M & y_00s, "Genre"].value_counts(normalize=True)
```

```python
df.loc[between_2M_10M & y_10s, "Genre"].value_counts(normalize=True).plot.bar()
```

```python
df.loc[~df["Genre"].isin(["Puzzle", "Strategy", "Adventure"])]
```

```python
# Let's do an exercise to see if you got this.
# Let's find out which Racing games had the highest sales in EU compared to the NA, JP, and Other Sales.
# between 2000-2010

racing = df["Genre"] == "Racing"
y_00s = df["Year"].between(2000, 2009)
highest_eu = (df["EU_Sales"] > df["JP_Sales"]) & (df["EU_Sales"] > df["NA_Sales"]) & (df["EU_Sales"] > df["Other_Sales"])
```

```python
df.loc[racing & highest_eu & y_00s, ["Name", "EU_Sales"]].sort_values(by="EU_Sales", ascending=False)
```

```python
df.groupby("Genre")["NA_Sales"].mean()
```
